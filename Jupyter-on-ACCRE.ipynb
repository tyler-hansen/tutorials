# Setting up system to use Bash and R in Jupyter notebooks on ACCRE


## 1. Load anaconda


    module load Anaconda3/5.0.1

## 2. Create virtual environment with conda


__I named my virtualenv 'jupyter'__

    conda create -n jupyter python=3.6

## 3. Activate conda environment


    conda activate jupyter

__To deactivate an active environment, use__

    conda deactivate

__To get info about conda environment__

    conda info --envs

        Should see this:

        conda environments:

        base                     /accre/arch/easybuild/software/BinDist/Anaconda3/5.0.1
        jupyter              *  /home/hansetj1/.conda/envs/jupyter
        * current env

## 4. Install relevant python packages


    python -m pip install --upgrade pip

    pip install ipykernel numpy scipy matplotlib pandas scikit-learn hdfs

## 5. Install bash kernel for jupyter


    pip install bash_kernel

    OR 

    python -m bash_kernel.install # should do the same thing as pip command above

## 6. Add bash kernel to jupyter


    python -m ipykernel install --user --name hansetj1 --display-name "Python (jupyter)"

## 7. Install R kernel for jupyter


    conda install -c r r-irkernel

## 8. Go to the R install within the virtual environment:


    cd /home/hansetj1/.conda/envs/jupyter/lib/R/bin

## 9. Add R kernel to jupyter 


__Launch R using the installed R binary, execute installspec function, then quit R:__

    R

    IRkernel::installspec()

    q()

## 10. Install relevant R packages through conda. Most bioconductor packages can be found on bioconda(https://anaconda.org/bioconda). 


__For example, to install DESeq2:__ 

    conda install -c bioconda bioconductor-deseq2

## 11. Launch jupyter notebook portal on ACCRE


__go to http://portal.accre.vanderbilt.edu__  
    -> interactive apps  
    -> Jupyter Notebook  
    -- Working Directory  

        example: /data/hodges_lab/ATAC-STARR_V2/bin

    -- Python Version

        python3.6.3/Anaconda 5.0.1

    -- Python or Conda Virtual Environment

        example: /home/hansetj1/.conda/envs/jupyter

__launch job__

## 12. Connect to Jupyter
